# Menyiapkan Data Heart disease

```python
import pandas as pd
import numpy as np
```

```python
# membaca dataset csv dari url
url = 'https://raw.githubusercontent.com/kb22/Heart-Disease-Prediction/master/dataset.csv'
data = pd.read_csv(url)
```

```python
data.head()
```

Penjelasan data heart disease

*   age - umur
*   sex - pria=1 wanita=0
*   cp - type sakit dada
*   trestbps - tekanan darah
*   chol - serum cholestoral mg/dl 
*   fbs - gula darah > 120 mg/dl (iya=1 tidak=0)
*   restecg - resting electrocardiographic results
*   thalach - maximum heart rate achieved
*   exang - exercise induced angina (1 = yes; 0 = no)
*   oldpeak - ST depression induced by exercise relative to rest
*   slope - the slope of the peak exercise ST segment
*   ca - number of major vessels (0-3) colored by flourosopy
*   thal - 3 = normal; 6 = fixed defect; 7 = reversable defect
*   target - hasil prediksi sakit=1 tidak=0 



## Implementasi klasifikasi KNN



memisahkan hasil prediksi dengan fitur

```python
# pilih semua kolom kecuali kolom terakhir
x = data.iloc[:, :-1]
# pilih hanya kolom target
y = data.loc[:, "target"]
y = data['target'].values
```

bagi data menjadi data train dan data test, 80:20

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

import model knn dari sklearn

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

coba model prediksi knn dengan data baru

```python
x_new = [[63,1,3,145,233,1,0,150,0,2.3,0,0,1]]
y_predict = knn.predict(x_new)

print(y_predict[0])
```
