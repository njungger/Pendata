# Implementasi K-Mean Clustering

```python
import pandas as pd
import numpy as np
from numpy import unique
from numpy import where
from matplotlib import pyplot
from sklearn import preprocessing
from sklearn.cluster import KMeans
```

## Menyiapkan Data

```python
# membaca dataset csv dari url
url = 'https://raw.githubusercontent.com/njungger/dataset/main/IRIS.csv'
data = pd.read_csv(url)
```

```python
data.head()
```

```python
# hapus kolom variety
data_without_label = data.drop(columns=["species"])
data_without_label
```

## Hitung Data

```python
# separate target 

# values
X=data.iloc[:,0:4].values

# 
y = data.values[:,4]
```

```python
le = preprocessing.LabelEncoder()
labels = le.fit_transform(y)
```

```python
labels
```

```python
# data_tes = np.array(labels)
kmeans = KMeans(n_clusters=3, n_init=1, max_iter=1)
kmeans.fit(X)
# kmeans.labels_
labels = kmeans.predict(X)

centroids = kmeans.cluster_centers_
```

```python
centroids
```

```python
data = pd.DataFrame(labels, columns=["Label"])
```
